```python
import requests
import pandas as pd

# API URL
url = "https://jsonplaceholder.typicode.com/posts"

# Fetch data
response = requests.get(url)

# Check status
print("Status Code:", response.status_code)

# Convert to JSON
data = response.json()

# Convert to DataFrame
df = pd.DataFrame(data)

# Show first 5 rows
print(df.head())

# Save to CSV
df.to_csv("posts_data.csv", index=False)

print("Saved as posts_data.csv")
```

```python
# Dataset info
print("\nDataset Info:")
print(df.info())

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Number of posts per user
print("\nPosts per User:")
print(df['userId'].value_counts())

# Title length
df['title_length'] = df['title'].apply(len)

print("\nAverage Title Length:", df['title_length'].mean())
```
